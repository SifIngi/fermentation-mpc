```python
# In the beginning of the batch no substrate inhibition occurs. From this early exponential phase with low concentration of substrate, we can estimate the mu_max and the Ks value. 
# Ks is the substrate saturation constant. 

# We have the relation 1/mu = 1/mu_max + Ks/mu_max*(1/S) and can therefore plot 1/mu against 1/S 

mu_biomass = [0.1186, 0.1790, 0.2131, 0.2131, 0.3385, 0.2865, 0.4010]
substrate_initial = [0.5, 1.0, 2.0, 3.0, 5.0, 7.5, 10.0]

mu_biomass_inverse = np.divide(1,mu_biomass)
substrate_initial_inverse = np.divide(1,substrate_initial)

slope, intercept, r_value, p_value, std_err = scipy.stats.linregress(substrate_initial_inverse, mu_biomass_inverse)
print(slope, intercept, r_value, p_value, std_err)

plt.plot(substrate_initial_inverse, mu_biomass_inverse, 'o', label='original data')
plt.plot(substrate_initial_inverse, intercept + slope*substrate_initial_inverse, 'r', label='fitted line')
plt.legend()
plt.show()

```

```python
# This means that the intercept = 1/mu_max = 2.9063
# Therefore mu_max is 
mu_max = 1/2.9063
print(mu_max)

# The slope = Ks/mu_max => 2.8120 = Ks/ mu_max => 2.8120*mu_max = Ks 
Ks = mu_max*2.8120
print(Ks)
```

```python
# At large substrate concentration inhibition occur. The Ki constant is determined by estimating the parameter, based on the mu data and the model

# the time here is the substrate concentration (this is just because tellurium and copasi calls the x axis for time, we just pretend that it is substrate, because mu is a function of substrate concentration)

mu_model = '''
model *IDModel()

    ######## Set the compartment to 1, otherwise it will be multiplied by the compounds. 
    compartment comp1;
    comp1 =1;


    ######## Specify the species in the compartment
    mu in comp1; #glucose in comp1;


    ######## Constants
    mu_max = 0.3441
    Ks = 0.9676
    Ki = 247.9697642 #175.62317475 # should be estimated

    ######## Initial conditions
    #glucose = 0.5 # [g/L]
    #biomass = 0.1
    #rate = 0.00004

    ######## Functions
    mu := mu_max*time/(time+Ks+(time^2/Ki)) # [1/h]
    
    ######## Mass Balances 
    #eq_glucose: -> glucose; rate# [c-mol/h]
    #eq_biomass: -> biomass; mu*biomass # [c-mol/h]


    end
'''

r = te.loada(mu_model)

r.exportToSBML('mu_model.xml')
r.timeCourseSelections = ['time', 'mu']
results = r.simulate(0.5, 150, 100)
```

```python
plt.figure(num=None, figsize=(11, 7), dpi=90, facecolor='w', edgecolor='k')

plt.plot(results[:, 0], (results[:, 1]))
plt.scatter(substrate_initial, mu_biomass)
plt.legend(['mu from model', 'mu from data'], loc='upper left')
plt.xlabel('Time (hours)')
plt.ylabel('Biomass (c-mole)')
```
