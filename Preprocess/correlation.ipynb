```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

matplotlib.rc('xtick', labelsize=22)     
matplotlib.rc('ytick', labelsize=22)
```

```python
# Putting values from correlation matrix that are provided from Copasi when making parameter estimation
data = {'$K_{i}$': [1, -0.0747, 0.0840, -0.2247, 0.0472], '$K_{s}$': [-0.0747, 1, 0.6272, 0.8841, 0.4978], '$K_{s_{qs}}$': [0.0839, 0.6272, 1, 0.5706, 0.9701], r'$\mu_{max}$': [-0.2247, 0.8841, 0.5706, 1, 0.5237], '$qs_{max}$': [0.0472, 0.4978, 0.9701, 0.5237, 1]}
pd.DataFrame.from_dict(data, orient='index')
corr = pd.DataFrame.from_dict(data, orient='index', columns=['$K_{i}$', '$K_{s}$', '$K_{s_{qs}}$', r'$\mu_{max}$', '$qs_{max}$'])
```

```python
# Make heatmap
f, ax = plt.subplots(figsize=(10, 8))
sns.heatmap(corr, mask=np.zeros_like(corr, dtype=np.bool), cmap=sns.diverging_palette(220, 10, as_cmap=True),
            square=True, ax=ax)

plt.title('Correlation plot of parameters', fontsize = 25)
plt.savefig('figures/parameters/correlationplot.pdf')

plt.show()
```
