```python
# Load and parse OD values
OD = 'OD/SER_C016_AMBR_ODMeasurements.xlsx'
OD = pd.ExcelFile(OD)
OD_R1 = OD.parse('R1')

# Last 3 rows are NAN's, so they are deleted
# Then converted into biomass
OD_R1 = OD_R1[:-3]
biomass_R1 = OD_R1['OD']*0.38
```

```python
print('Substrate conc = 0.5 g/L')
plot_biomass_R1 = 'biomass_R1'
plot_biomass(sample_timepoints_R1_to_R8['R1 Time (hours)'], biomass_R1, plot_biomass_R1)
semilog_plot_biomass(sample_timepoints_R1_to_R8['R1 Time (hours)'], biomass_R1)

# From this we can observe that there must be a straight line from index 3-8 for biomass. 
# Then we can make a linear regression on this - and the slope is then maximum growth rate 

# Choose points from exponential phase
biomass_R1_exponential_phase = biomass_R1[3:9]

# Take the logarithmic of the exponential phase and save the timepoints that corresponds to the data
biomass_R1_exponential_phase = np.log(biomass_R1_exponential_phase)
sample_timepoints_R1_exponential_phase = sample_timepoints_R1_to_R8['R1 Time (hours)'][3:9]

# Find the slope from the linear regression on the exponential phase
lin_reg_slope_mu(sample_timepoints_R1_to_R8['R1 Time (hours)'], biomass_R1, biomass_R1_exponential_phase, sample_timepoints_R1_exponential_phase, plot_biomass_R1)

# Find the maximum growth rate from CER
data_frame_selected_values_R1, mu_R1 = calculate_mu(R1, 1)
print('mumax from CER:',np.amax(mu_R1[19:28]*60))
```
