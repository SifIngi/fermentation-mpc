## Linear regression using 95 % confidence interval on the qs values until 19 g/L initial substrate

```python
qs = (1/qs_values['qs (gsub/gbio*h)'][:8])
substrate_initial = (1/qs_values['Substrate values g/L'][:8])


x = substrate_initial
y = qs


def f(x, a, b):
    return a*x + b


# The following returns p from the fitted curve which consists of the intercept and slope 
# and returns the covarians matrix. The function curve_fit is from scipy optimize.
p, pcov = curve_fit(f, x, y)

# The diagonals in covariance is the variance of the parameter estimation. 
# Then we can compute the standard deviation on the parameters with ->
perr = np.sqrt(np.diag(pcov))

# I create evenly spaced numbers from the minimum value of x to the maximum value of x. 
# and ps are samples that are drawn from a normal multivariate distribution around the mean (which is the slope and intercept), and covariance matrix (). 
# Then creates samples around the intercept and the slope value. 
xi = linspace(np.min(x), np.max(x), 100)
ps = np.random.multivariate_normal(p, pcov, 10000)

# This creates (10000,100) array i.e. 100 y samples pr. set of slope and intercept sample, so 100 samples 10000 times. 
# In percentile axis 0 means it takes the values in the column, order it from lowest to highest then multiply 0.025*10.000 test scores for the lower confidence interval. 
ysample = np.asarray([f(xi, *pi) for pi in ps])
lower = np.percentile(ysample, 2.5, axis=0)
upper = np.percentile(ysample, 97.5, axis=0)


plt.figure(num=None, figsize=(11, 7), dpi=90, facecolor='w', edgecolor='k')
plt.plot(x, y, 'bo')
plt.plot(x, p[1] + p[0]*x, 'r-')
plt.plot(xi, lower, 'b--')
plt.plot(xi, upper, 'b--')
plt.ylabel('1/qs',fontsize = 13)
plt.xlabel('1/[S]',fontsize = 13)
plt.title('1/qs vs 1/[S]',fontsize = 20)
plt.savefig('figures/kinetics/invqs_vs_invsub.png')
plt.show()


# Then I delete the point from reactor 4 from the outlier but I also delete point 6 since that growth rate was also classified as an outlier and the qs was calculated from the growth rate
qs_values = qs_values.drop(qs_values.index[[3,5]])
qs = (1/qs_values['qs (gsub/gbio*h)'][:6])
substrate_initial = (1/qs_values['Substrate values g/L'][:6])

x = substrate_initial
y = qs

# Makes a linear function
def f(x, a, b):
    return a*x + b
p, pcov = curve_fit(f, x, y)

print(p)
plt.figure(num=None, figsize=(11, 7), dpi=90, facecolor='w', edgecolor='k')
plt.plot(x, y, 'bo')
plt.plot(x, p[1] + p[0]*x, 'r-')
plt.plot(xi, lower, 'b--')
plt.plot(xi, upper, 'b--')
plt.ylabel('1/mu',fontsize = 13)
plt.xlabel('1/[S]',fontsize = 13)
plt.title('1/mu vs 1/[S]',fontsize = 20)
plt.show()

# Calculates R^2 value
SSE = np.sum((y - (p[1] + p[0]*x))**2)
SST = np.sum((y - mean(y))**2)
r_squared = 1 - (SSE/SST)
print(r_squared)
print(p[1])

```

```python
# This means that the intercept = 1/qs_max 
# Therefore qs_max is 
qs_max = 1/p[1] 
print(qs_max)

# The slope = Ks/qs_max =>  qs_max*slope = Ks
Ks = qs_max*p[0] 
print(Ks) 
```

```python
qs = (qs_values['qs (gsub/gbio*h)'][:6])
substrate_initial = (qs_values['Substrate values g/L'][:6])

x = substrate_initial
y = qs

def func(x):
    return  qs_max*x/((x+Ks))

xi = linspace(np.min(x), np.max(x), 100)
plt.figure(num=None, figsize=(11, 7), dpi=90, facecolor='w', edgecolor='k')
plt.plot(x, y, 'bo')
plt.plot(xi, func(xi), 'r-')
plt.ylabel('qs (gsub/(gbio*h))',fontsize = 13)
plt.xlabel('Initial substrate (glucose) concentration g/L',fontsize = 13)
plt.title('qs vs substrate',fontsize = 20)
plt.savefig('figures/kinetics/qs_vs_sub_model.png')
plt.show()
```
