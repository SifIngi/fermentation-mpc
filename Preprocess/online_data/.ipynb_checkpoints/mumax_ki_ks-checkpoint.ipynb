## Coefficient from linear regression using 95 % confidence interval  

```python
# New conf way
from pylab import *
from scipy.optimize import curve_fit

mu_biomass = [0.1186, 0.1790, 0.2131, 0.3385, 0.4010]
substrate_initial = [0.5, 1.0, 2.0, 5.0, 10.0]

mu_biomass = np.divide(1,mu_biomass)
substrate_initial = np.divide(1,substrate_initial)

# Lambda is a oneline function, lambda is good to use when we do not wont to run the function more than one time. (ikke afskrift)
# They are like normal functions and behave like them (afskrift). Substrate_initial is the argument as well as p. (everthing before :)

f = lambda substrate_initial, *p: polyval(p, substrate_initial)

# p is the parameters, p = intercept, slope. 
# Find the covariance of the parameters 
p, cov = curve_fit(f, substrate_initial, mu_biomass, [1, 1])

x = substrate_initial
y = mu_biomass

xi = linspace(np.min(x), np.max(x), 100)
ps = np.random.multivariate_normal(p, cov, 10000)
ysample = np.asarray([f(xi, *pi) for pi in ps])
lower = percentile(ysample, 2.5, axis=0)
upper = percentile(ysample, 97.5, axis=0)

# regression estimate line
y_fit = poly1d(p)(xi)

# plot
plot(x, y, 'bo')
plot(xi, y_fit, 'r-')
plot(xi, lower, 'b--')
plot(xi, upper, 'b--')

print(p)

# Then we can remove samples that are higher than upper bound or lower than lower bound. To check if we get the same parameters
print(mu_biomass)
print(substrate_initial)

# We delete last point from array, 4, 
```

## Linear regression based on manually deleting data that does not fit

```python
# We have the relation 1/mu = 1/mu_max + Ks/mu_max*(1/S) and can therefore plot 1/mu against 1/S 

mu_biomass = [0.1186, 0.1790, 0.2131, 0.3385, 0.4010]
substrate_initial = [0.5, 1.0, 2.0, 5.0, 10.0]

mu_biomass_inverse = np.divide(1,mu_biomass)
substrate_initial_inverse = np.divide(1,substrate_initial)

slope, intercept, r_value, p_value, std_err = scipy.stats.linregress(substrate_initial_inverse, mu_biomass_inverse)
print(slope, intercept, r_value, p_value, std_err)

plt.plot(substrate_initial_inverse, mu_biomass_inverse, 'o', label='original data')
plt.plot(substrate_initial_inverse, intercept + slope*substrate_initial_inverse, 'r', label='fitted line')
plt.legend()
plt.show()

# This is pretty close to the former found parameter. The slope could vary a bit
```

## Estimating Ki based on linear regression with removed points that are higher than upper bound or lower than lower bound based on 95 % confidence interval

```python
mu_biomass = [0.1186, 0.1790, 0.3385, 0.4010, 0.3304, 0.3031, 0.2656, 0.1916]
substrate_initial = [0.5, 1.0, 5.0, 10.0, 20.0, 40.0, 60.0, 150.0]
```

```python
# At large substrate concentration inhibition occur. The Ki constant is determined by estimating the parameter, based on the mu data and the model

# the time here is the substrate concentration (this is just because tellurium and copasi calls the x axis for time, we just pretend that it is substrate, because mu is a function of substrate concentration)

mu_model = '''
model *IDModel()

    ######## Set the compartment to 1, otherwise it will be multiplied by the compounds. 
    compartment comp1;
    comp1 =1;


    ######## Specify the species in the compartment
    mu in comp1; #glucose in comp1;


    ######## Constants
    mu_max = 0.434
    Ks = 1.349
    Ki = 247.9697642 #175.62317475 # should be estimated

    ######## Initial conditions
    #glucose = 0.5 # [g/L]
    #biomass = 0.1
    #rate = 0.00004

    ######## Functions
    mu := mu_max*time/(time+Ks+(time^2/Ki)) # [1/h]
    
    ######## Mass Balances 
    #eq_glucose: -> glucose; rate# [c-mol/h]
    #eq_biomass: -> biomass; mu*biomass # [c-mol/h]


    end
'''

r = te.loada(mu_model)

r.exportToSBML('model_and_estimation/mu_model.xml')
r.timeCourseSelections = ['time', 'mu']
results = r.simulate(0.5, 150, 100)
```

```python
plt.figure(num=None, figsize=(11, 7), dpi=90, facecolor='w', edgecolor='k')

plt.plot(results[:, 0], (results[:, 1]))
plt.scatter(substrate_initial, mu_biomass)
plt.legend(['mu from model', 'mu from data'], loc='upper left')
plt.xlabel('Time (hours)')
plt.ylabel('mu Biomass (c-mole)')
```

```python
# Write the data to a csv file that can be used in copasi

mu_biomass = pd.DataFrame(mu_biomass)
substrate_initial = pd.DataFrame(substrate_initial)


substrate_mu = substrate_initial
substrate_mu['mu'] = mu_biomass

substrate_mu.to_csv('model_and_estimation/substrate_mu.csv')
```
