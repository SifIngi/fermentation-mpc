```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
mu_max_CO2 = [0.6331, 0.6165, 0.5764, 0.7083, 0.7826, 0.6860, 0.5581, 0.6398, 0.5226, 0.4683, 0.4386, 0.3347]
mu_max_biomass = [0.1186, 0.1790, 0.2131, 0.2131, 0.3385, 0.2865, 0.4010, 0.3304, 0.3031, 0.2656, 0.9483, 0.1916]
mu_max_CO2 = [i*0.38 for i in mu_max_CO2]
substrate_initial = [0.5, 1.0, 2.0, 3.0, 5.0, 7.5, 10.0, 20.0, 40.0, 60.0, 120.0, 150.0]
```

```python
plt.figure(1,figsize=(20, 7))
plt.scatter(substrate_initial,mu_max_biomass)
plt.scatter(substrate_initial,mu_max_CO2)
plt.legend(['mu from biomass', 'mu from CO2'], loc='upper right')
plt.ylabel('mu_max (1/h)', fontsize = 20)
plt.xlabel('substrate concentration (g/L)', fontsize = 20)
plt.title('mu max from experiments')

plt.show()

plt.scatter(substrate_initial,mu_max_biomass)
plt.show()
```

```python
print(mu_max_biomass)


model = LinearRegression(fit_intercept=True)
model.fit(np.array([0.6398, 0.5226, 0.4683, 0.3347])[:, np.newaxis],np.array([0.3304, 0.3031, 0.2656, 0.1916]))



xfit = np.linspace(0.33, 0.64, 1000)
yfit = model.predict(xfit[:, np.newaxis])

plt.scatter([0.6398, 0.5226, 0.4683, 0.3347], [0.3304, 0.3031, 0.2656, 0.1916])
plt.plot(xfit, yfit)
plt.show()



print("Model slope (mu_max):    ", model.coef_[0])


```
